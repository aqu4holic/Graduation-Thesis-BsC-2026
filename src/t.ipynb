```python
import matplotlib.pyplot as plt
import numpy as np
def plot_radial_nguhanh(elements: dict, title="Biểu đồ khí ngũ hành"):
    hanh_labels = ["Mộc", "Hỏa", "Thổ", "Kim", "Thủy"]
    values = [elements.get(h, 0) for h in hanh_labels]
    values += values[:1]  # khép kín vòng tròn
    angles = np.linspace(0, 2 * np.pi, len(hanh_labels), endpoint=False).tolist()
    angles += angles[:1]
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    ax.plot(angles, values, linewidth=2, linestyle='solid')
    ax.fill(angles, values, alpha=0.3)
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    ax.set_thetagrids(np.degrees(angles[:-1]), hanh_labels)
    ax.set_title(title, size=14, fontweight='bold')
    ax.grid(True)
    plt.tight_layout()
    plt.show()
if __name__ == "__main__":
    sample = {"Mộc": 1, "Hỏa": 1, "Thổ": 2, "Kim": 1, "Thủy": 3}
    plot_radial_nguhanh(sample)
```
